```python
import os
import cv2
import numpy as np
import torch
import torch.nn as nn
from torch.nn.functional import log_softmax
from matplotlib import pyplot as plt

# Device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Label map (make sure this matches training)
base_dir = r"C:\Users\Asus-2024\Downloads\Final_proper_dataset_updated\Final_dataset\customized_train"
categories = ['Consonants', 'Numerals', 'mansample-final', 'uyirezhuthukkal']

label_names = []
for category in categories:
    category_path = os.path.join(base_dir, category)
    for subfolder in os.listdir(category_path):
        subfolder_path = os.path.join(category_path, subfolder)
        if os.path.isdir(subfolder_path) and subfolder not in label_names:
            label_names.append(subfolder)

idx2char = {i: name for i, name in enumerate(label_names)}
  # Fill this with the same label list from training
idx2char = {i: name for i, name in enumerate(label_names)}

# Model Definition (same as training)
class CNN_BiLSTM_CTC(nn.Module):
    def __init__(self, num_classes):
        super(CNN_BiLSTM_CTC, self).__init__()
        self.cnn = nn.Sequential(
            nn.Conv2d(1, 64, 3, padding=1), nn.BatchNorm2d(64), nn.ReLU(), nn.MaxPool2d((2, 2)),
            nn.Conv2d(64, 128, 3, padding=1), nn.BatchNorm2d(128), nn.ReLU(), nn.MaxPool2d((2, 2)),
            nn.Conv2d(128, 256, 3, padding=1), nn.BatchNorm2d(256), nn.ReLU(), nn.MaxPool2d((2, 2))
        )
        self.lstm = nn.LSTM(input_size=1024, hidden_size=512, num_layers=3,
                            bidirectional=True, batch_first=True)
        self.dropout = nn.Dropout(0.25)
        self.fc = nn.Linear(1024, num_classes + 1)

    def forward(self, x):
        x = self.cnn(x)
        x = x.permute(0, 3, 1, 2)
        x = x.reshape(x.size(0), x.size(1), -1)
        x, _ = self.lstm(x)
        x = self.dropout(x)
        x = self.fc(x)
        x = x.permute(1, 0, 2)
        return log_softmax(x, dim=2)

# Load model
model = CNN_BiLSTM_CTC(num_classes=len(label_names)).to(device)
model.load_state_dict(torch.load("cnn_bilstm_ctc_grantha_150.pth", map_location=device))
model.eval()

# Greedy Decoder
def greedy_decoder(output):
    pred = output.permute(1, 0, 2)
    pred = torch.argmax(pred, dim=2)
    decoded = []
    for p in pred:
        seq = []
        prev = -1
        for char in p:
            if char.item() != prev and char.item() != len(label_names):
                seq.append(char.item())
            prev = char.item()
        decoded.append(seq)
    return decoded

# Inference on a single image
def predict_image(image_path):
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    img = cv2.resize(img, (128, 32))
    img = img.astype(np.float32) / 255.0
    img = torch.tensor(img).unsqueeze(0).unsqueeze(0)  # [1, 1, 32, 128]
    img = img.to(device)

    with torch.no_grad():
        output = model(img)
        pred_indices = greedy_decoder(output)[0]
        pred_labels = [idx2char[idx] for idx in pred_indices]
        return pred_labels

# Example usage
test_image_path = r"C:\Users\Asus-2024\OneDrive\Pictures\Screenshots\Screenshot 2025-04-25 075344.png"
prediction = predict_image(test_image_path)
pred_unicode = [Label_Dict[label] for label in prediction]
print("Predicted Output:", ''.join(pred_unicode))
#print("Predicted sequence:", Label_Dict[prediction])

# Optional: visualize the image with prediction
img = cv2.imread(test_image_path, cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')
#plt.title(f"Prediction: {' '.join(prediction)}")
plt.axis('off')
plt.show()
```
